```python
import yfinance as yf
import pandas as pd

def fetch_stock_data(symbol, start_date, end_date):
    try:
        df = yf.download(symbol, start=start_date, end=end_date, interval="1d")
        return df[['Close']]
    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")
        return pd.DataFrame()

def add_next_day_increase_column(df, percentage_threshold=0.5):
    # Shift the 'Close' column by one day to get the next day's closing price
    df['Next_Day_Close'] = df['Close'].shift(-1)

    # Create a new column indicating whether the closing price increased by a certain percentage on the next day
    df['Close_Increase'] = (df['Next_Day_Close'] / df['Close'] - 1) > (percentage_threshold / 100)

    # Drop the 'Next_Day_Close' column as it's no longer needed
    df = df.drop(columns=['Next_Day_Close'])

    return df

# Example usage
symbol = 'AAPL'  # Replace with the desired stock symbol
start_date = '2022-01-01'
end_date = '2023-01-01'

percentage_threshold = 0.5  # Adjust the percentage threshold as needed

stock_data = fetch_stock_data(symbol, start_date, end_date)

if not stock_data.empty:
    stock_data_with_increase = add_next_day_increase_column(stock_data, percentage_threshold)

    # Save the DataFrame to a CSV file
    csv_file_path = f'{symbol}_closing_data_with_increase.csv'
    stock_data_with_increase.to_csv(csv_file_path, index=False)

    print(f"Closing data with next day increase saved to {csv_file_path}")
```
